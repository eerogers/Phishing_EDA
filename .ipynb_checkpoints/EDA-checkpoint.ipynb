```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import os
import seaborn as sns
from sklearn.decomposition import PCA
from sklearn.preprocessing import scale
#from library.sb_utils import save_file

data = pd.read_csv('phishing_data.csv')
class1 = data[data['CLASS_LABEL'] == 0] #Class one, I believe, describes phishing websites; '0' corresponds to 
#'False' and the original csv was named 'Legitimate websites' - This is my reasoning
class2 = data[data['CLASS_LABEL'] == 1]
data_columns = data.columns
data_index = data.index
class_index = data.set_index('CLASS_LABEL')
print(data_columns)
print(data.groupby(by='CLASS_LABEL').mean())
```

```python
scaled_data = scale(data)
scaled_data_df = pd.DataFrame(scaled_data, columns=data_columns)
scaled_data_df.head() #note: not grouped by class_label
scaled_data_df.mean()
```

```python
scaled_data_df.std(ddof=0)
```

```python
data_pca = PCA().fit(scaled_data_df)
plt.plot(data_pca.explained_variance_ratio_.cumsum())
plt.xlabel('Component #')
plt.ylabel('Cumulative ratio variance')
plt.title('Cumulative variance ratio explained by PCA components for phishing website statistics');
#I think it's notable below that this doesn't capture a large percentage of the variance within a short span
```

```python
data_pca_x = data_pca.transform(scaled_data_df)
data_pca_x.shape
```

```python
x = data_pca_x[:, 0]
y = data_pca_x[:, 1]
pc_var = 100 * data_pca.explained_variance_ratio_.cumsum()[1]
plt.subplots(figsize=(10,8))
plt.scatter(x=x, y=y, c=data['CLASS_LABEL'])
plt.xlabel('First component')
plt.ylabel('Second component')
plt.title(f'Phishing website summary PCA, {pc_var:.1f}% variance explained')
#for s, x, y in zip(data, x, y):
#    plt.annotate(s, (x, y))
```

```python
#I'm not sure how much of that went correctly - I'm going to try again with just the float values and see what happens.
pct_labels = ['PctExtHyperlinks', 'PctExtResourceUrls', 'PctNullSelfRedirectHyperlinks']
#pct_df = data[pct_labels]
transposed_pct = data[pct_labels].transpose()
scaled_pct = scale(data[pct_labels])
scaled_pct_df = pd.DataFrame(scaled_pct, columns=pct_labels)
scaled_pct_df.head() #note: not grouped by class_label
scaled_pct_df.mean()
```

```python
scaled_pct_df.std(ddof=0)
```

```python
pct_PCA = PCA().fit(scaled_pct_df)
plt.subplots(figsize=(10, 6))
plt.plot(pct_PCA.explained_variance_ratio_.cumsum())
plt.xlabel('Component #')
plt.ylabel('Cumulative ratio variance')
plt.title('Cumulative variance ratio explained by PCA components for state/resort summary statistics');
```

```python
pct_pca_x = pct_PCA.transform(scaled_pct_df)
pct_pca_x.shape #I'm not sure I should be doing what I did in the guided capstone because I end up using 
#all the individual points - not sure if that's right but maybe?- Labeling those points seems confusing though.
```

```python
x = pct_pca_x[:, 0]
y = pct_pca_x[:, 1]
columns = data_index
pc_var = 100 * pct_PCA.explained_variance_ratio_.cumsum()[1]
plt.subplots(figsize=(10,8))
plt.scatter(x=x, y=y, c=data['CLASS_LABEL'])
plt.xlabel('First component')
plt.ylabel('Second component')
plt.title(f'Ski states summary PCA, {pc_var:.1f}% variance explained')
#for s, x, y in zip(columns, x, y):
#    plt.annotate(s, (x, y))
```
